```python
# 일별 대여건수 1-6 이용량 변화 추이선 선 그래프 표현

# 한글 깨짐 문제 해결
import pandas as pd
df = pd.read_csv('서울특별시 공공자전거 일별 대여건수_24.1-6.csv', encoding='cp949')
df.head()
```

```python
# 컬럼명 영어로 변경
df = df.rename(
    columns={'대여일자' : 'rental_date',
             '대여건수' : 'rental_number'}
)
df.head()
```

```python
# 데이터 정보 확인
df.info()
```

```python
# 선 그래프
# x축 연월일 나타낸 문자 가로에 여러 번 겹쳐서 굵은 선으로 보이게 됨
import seaborn as sns

sns.lineplot(data=df, x='rental_date', y='rental_number')
```

```python
# rental_date 문자열 타입 -> 날짜 시간 타입 
df['rental_date'] = pd.to_datetime(df['rental_date'])

# 변수 타입 다시 확인
df.info()
```

```python
# 월만 추출 (월 변수 생성)
df['month'] = df['rental_date'].dt.month
df['month']
```

```python
# x축에 월만 추출한 선그래프
sns.lineplot(data=df, x='month', y='rental_number')
```
